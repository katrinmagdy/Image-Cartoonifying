# **Imports**

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import pandas as pd
import csv
import math
from matplotlib import cm
from PIL import Image, ImageFilter
```

# **Test1**

# **Read Image**

```python
img = cv2.imread('/content/example.jpeg')
print(img.shape)
```

```python
cv2_imshow(img)
```

# **Read Grayscale Image**

```python
gray_img = cv2.imread('/content/example.jpeg', 0)
cv2_imshow(gray_img)
```

# **Noise Reduction Using Median Filter**

```python
median_img = cv2.medianBlur(gray_img,5)
cv2_imshow(median_img)
```

# **Edge Detection Using Laplacian Filter**

```python
laplacian_img = cv2.Laplacian(median_img,0, ksize=3)
cv2_imshow(laplacian_img)
```

# **Inverse Binary Thresholding**

```python
histr,_ = np.histogram(laplacian_img,bins=256)
plt.plot(histr)
plt.show()
```

```python
_,binary_img = cv2.threshold(laplacian_img, 45, 255,cv2.THRESH_BINARY_INV)
plt.imshow(binary_img, 'gray')
```

# **Generating a color painting and a cartoon (Applying Bilateral Filter)**

```python
x=6
y=6
blur = cv2.bilateralFilter(img,21,x,y)
blur = cv2.bilateralFilter(blur,21,x,y)
blur = cv2.bilateralFilter(blur,21,x,y)
blur = cv2.bilateralFilter(blur,21,x,y)
blur = cv2.bilateralFilter(blur,21,x,y)
blur = cv2.bilateralFilter(blur,21,x,y)


cv2_imshow(blur)
```

# overlay the edge mask ”sketch” onto the bilateral filter ”painting”
using a black background and copying the ”painting” pixels that aren’t edges in the ”sketch” mask
#**The Final Output**

```python
output= np.zeros(((206, 334, 3)),dtype=int)
for i in range (206):
  for j in range(334):
    if binary_img[i,j] == 255:
      output[i,j] = blur[i,j]

cv2_imshow(output)
```
